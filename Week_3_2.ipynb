### **Segmenting and Clustering Neighborhoods in Toronto (Second Step)**

##### **Import Libraries and Data from First Step**

```python
import pandas as pd
import numpy as np

df_toronto = pd.read_csv('toronto_part1.csv')
df_toronto = df_toronto.drop(['Unnamed: 0'], axis=1)
df_toronto.head(10)
```

##### **Part 1: Get Geocodes**

```python
df_geodata = pd.read_csv("http://cocl.us/Geospatial_data")
df_geodata.head(10)
```

##### **Part 2: Match and Merge Geocodes**

```python
df_2 = pd.merge(df_toronto, df_geodata, left_on='Postcode', right_on='Postal Code')
df_2 = df_2.drop(["Postal Code"], axis=1)
df_2.head()
```

##### **Part 3: Save as CSV for Third Step**

```python
df_2.to_csv('toronto_part2.csv')
```
